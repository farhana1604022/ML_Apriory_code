```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("1604022_Depression and Happiness Factor.csv") 
```

```python
df.shape
```

```python
df=df.drop('Timestamp',axis=1)
```

```python
df
```

```python
#Preprocessing
#Making a list of data
info = df.to_numpy().tolist()
info
```

```python
for values in info:
    for i in range(len(values)):
        s = ''
        if type(values[i]) != str:
            s = str(values[i])
            values[i] = s
```

```python
pip install mlxtend
```

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
tr = TransactionEncoder()
t_info = tr.fit(info).transform(info)
t_info
```

```python
df = pd.DataFrame(t_info,columns=tr.columns_) # create a dataframe
df
```

```python
from mlxtend.frequent_patterns import apriori
frequent_items = apriori(df, min_support=0.60, use_colnames=True)
frequent_items
```

```python

from mlxtend.frequent_patterns import association_rules
result = association_rules(frequent_items, metric='confidence', min_threshold=.50)
```

```python
new_result = result.loc[:, ['antecedents', 'consequents', 'support','confidence']]
new_result
```
